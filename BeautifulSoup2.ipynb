```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
link='https://forecast.weather.gov/MapClick.php?lat=37.7771&lon=-122.4196#.YxoH-nZByM8'
html_text=requests.get(link).text
bs=BeautifulSoup(html_text,'lxml')
```

```python
temperature=''

temprature_html=bs.find('p',attrs={"class":"myforecast-current-lrg"})
temperature=temprature_html.find(text=True).strip()
```

```python
html = bs.find('div',attrs={"id":"seven-day-forecast-container"})
h=html.find_all('li') 
```

```python
days=[]
high_low=[]
description=[]
short_description=[]
dictionary={}

for i in h:

    
    d=i.find('p',attrs={"class":"period-name"})
    days.append(d.find(text=True).strip())

    
   
    ds=i.find('div',attrs={"class":"tombstone-container"})
    more=(ds.find('img'))
    description.append(more['alt'])

    
    
    short_d=i.find('p',attrs={"class":"short-desc"})
    short_description.append(short_d.find(text=True).strip())

    
    
    try:
        highlow=i.find('p',attrs={"class":"temp temp-high"})
        high_low.append(highlow.find(text=True).strip())
    except AttributeError:
        highlow=i.find('p',attrs={"class":"temp temp-low"})
        high_low.append(highlow.find(text=True).strip())
    
```

```python
dictionary['Days']=days
dictionary['Description']=description
dictionary['Short Description']=short_description
dictionary['High/Low']=high_low
```

```python
df=pd.DataFrame.from_dict(dictionary)
df
```

```python
df.to_csv('result2.csv', index=False)
```
