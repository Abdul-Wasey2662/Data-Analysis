```python
from bs4 import BeautifulSoup

import requests
```

```python
url='https://www.passiton.com/inspirational-quotes'
```

```python
html_text=requests.get(url).text
```

```python
soup=BeautifulSoup(html_text,'lxml')
```

```python
table = soup.find('div',attrs={"class":"col-8 d-flex flex-column justify-content-top text-center"})
```

```python
t=table.find_all('a')
q_link=[]
q=[]
cat=[]
img_url=[]

# quoute link
for i in t:
    q_link.append(i['href'])
    
print(len(q_link))
z=table.find_all('img')

# quotes & image url
for a in z:
    q.append(a['alt'])
    img_url.append(a['src'])
    
z=table.find_all('h5')
# Category 
for f in z:
    cat.append(f.find(text=True).strip())

    

```

```python
# removing duplicate lniks
q_link = list(set(q_link)) 
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
scraped_data=[]
list_dict={}
# making dictionary to insert in csv
list_dict['category']=cat
list_dict['img']=img_url
list_dict['lines']=q
list_dict['url']=q_link
```

```python
df=pd.DataFrame.from_dict(list_dict)
df
```

```python
df.to_csv('file.csv', index=False)
```
